# Exploración de Datos del Experimento: Oddball Auditivo/Lingüístico

Este notebook realiza una exploración básica para validar la integridad de los datos experimentales generados por PsychoPy.

**Condiciones a verificar:**
1. **Frecuencia de Estímulos:** La palabra "MESA" (estímulo *deviant* u *oddball*) debería aparecer aproximadamente el 20% de las veces.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Cargar la data
dt = pd.read_csv('../data/634466_ling_auditory_oddball_2025-11-20_01h40.55.105.csv')

# Filtrar solo las filas que corresponden a los ensayos (donde se mostró un estímulo)
trials = dt.dropna(subset=['textStim']).copy()

print(f"Total de ensayos registrados: {len(trials)}")
trials.head()
```

## Análisis de Frecuencia de Estímulos
Verificamos la cantidad de veces que aparece cada palabra y su proporción.

```python
# Conteo absoluto
conteo = trials['textStim'].value_counts()
print("Conteo de estímulos:")
print(conteo)

print("\n---")

# Proporción (Porcentaje)
proporcion = trials['textStim'].value_counts(normalize=True)
print("Proporción de estímulos (Esperado: MESA ~ 0.20):")
print(proporcion)
```

```python
# Visualización rápida de la proporción
proporcion.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Distribución de Estímulos (MASA vs MESA)')
plt.ylabel('Proporción')
plt.axhline(y=0.2, color='r', linestyle='--', label='Objetivo (20%)')
plt.legend()
plt.show()
```
